```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
images = 'datasets/lego/images/'
```

```python
i=0
name=os.listdir(images)[i]
img = cv2.imread(images+name)

plt.imshow(img)
```

```python
s=img.shape
x=np.sort(np.random.randn(2)*[s[0]]/5+s[0]/2).astype(int)
y=np.sort(np.random.randn(2)*[s[1]]/5+s[0]/2).astype(int)

print(x[0],y[0])


c=np.random.randint(0,255,3)

max=200

x[1]=np.min([x[1],x[0]+max])
y[1]=np.min([y[1],y[0]+max])

img[x[0]:x[1],y[0]:y[1]]=c

print(c)

plt.imshow(img)
```

```python
for image in os.listdir(images):
    
    img = cv2.imread(images+image)

    s=img.shape
    x=np.sort(np.random.randn(2)*[s[0]]/5+s[0]/2).astype(int)
    y=np.sort(np.random.randn(2)*[s[1]]/5+s[0]/2).astype(int)

    c=np.random.randint(0,255,3)
    x[1]=np.min([x[1],x[0]+max])
    y[1]=np.min([y[1],y[0]+max])

    img[x[0]:x[1],y[0]:y[1]]=c

    cv2.imwrite(images+image, img)
    print(image)
```
