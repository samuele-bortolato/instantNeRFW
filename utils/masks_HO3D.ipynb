```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
masks_in = 'C:/Users/Sam/Downloads/HO3D_v3/HO3D_v3/HO3D_v3_segmentations_rendered/train/ABF10/seg'
masks_out = 'D:/Scuola/02_UniBo/2_anno/ML4CV/instantNeRFW/datasets/ABF10/masks'

if not os.path.exists(masks_out):
    os.makedirs(masks_out)
```

```python
for file in os.listdir(masks_in):
    print(file)
    img = cv2.imread(os.path.join(masks_in,file),cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img,(640, 480),interpolation=cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(masks_out,file),((img==149)*255).astype(np.uint8))
```
