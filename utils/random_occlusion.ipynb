```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
images = '../datasets/fox/images/'
```

```python
i=0
name=os.listdir(images)[i]
img = cv2.imread(images+name)

plt.imshow(img)
```

```python
s=img.shape
x=np.random.randint(s[0]/10, s[0]/4)
y=np.random.randint(s[0]/10, s[1]/4)
x_off=np.random.randint(0, s[0]-x)
y_off=np.random.randint(0, s[1]-y)

c=np.random.randint(0,255,3)

img[x_off:x_off+x,y_off:y_off+y]=c

print(c)

plt.imshow(img)
```

```python
for image in os.listdir(images):
    
    img0 = cv2.imread(images+image)
    img1 = cv2.imread(images+image)

    s=img0.shape
    x=np.random.randint(s[0]/10, s[0]/4)
    y=np.random.randint(s[0]/10, s[1]/4)
    x_off=np.random.randint(0, s[0]-x)
    y_off=np.random.randint(0, s[1]-y)
    c=np.random.randint(0,255,3)
    img0[x_off:x_off+x,y_off:y_off+y]=c

    x=np.random.randint(s[0]/10, s[0]/4)
    y=np.random.randint(s[0]/10, s[1]/4)
    x_off=np.random.randint(0, s[0]-x)
    y_off=np.random.randint(0, s[1]-y)
    c=np.random.randint(0,255,3)
    img1[x_off:x_off+x,y_off:y_off+y]=c

    path = image.split('.')[0] + '_.' + image.split('.')[1]
    cv2.imwrite(images+image, img0)
    cv2.imwrite(images+path, img1)
    print(image)
```
